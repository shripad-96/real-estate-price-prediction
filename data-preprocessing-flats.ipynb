```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('flats.csv',encoding='latin1')
df.sample(5)
```

```python
# shape
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated()
```

```python
df.shape
```

```python
# check for missing values
df.isnull().sum()
```

```python
# Columns to drop -> property_name, link, property_id
df.drop(columns=['link','property_id'], inplace=True)
```

```python
df.head()
```

```python
# rename columns
df.rename(columns={'area':'price_per_sqft'},inplace=True)
```

```python
df.head()
```

```python
# society
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society'].value_counts()
```

```python
df.head()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.replace('?','')

df.head()
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom'] = pd.to_numeric(df['bedRoom'].str.split(' ').str.get(0), errors='coerce')
```

```python
df.head()
```

```python
df['bathroom'] = pd.to_numeric(df['bathroom'].str.split(' ').str.get(0), errors='coerce')
```

```python
df.head()
```

```python
df['balcony'] = pd.to_numeric(df['balcony'].str.split(' ').str.get(0), errors='coerce')
```

```python
df.head()
```

```python
df['additionalRoom'].value_counts().shape
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df.head()
```

```python
# facing
df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['facing'].value_counts()
```

```python
df.insert(loc=1,column='property_type',value='flat')
```
