```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?â˜…', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df.head()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('â‚¹','').str.replace(',','').str.replace('?','')
```

```python
df['bathroom'] =pd.to_numeric(df['bathroom'].str.split(' ').str.get(0),errors='coerce')
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['noOfFloor'] = df['noOfFloor'].str.split(' ').str.get(0)
```

```python
df.head()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df.dropna(subset=['price_per_sqft'])
```

```python
def clean_price(price):
    try:
        # Remove 'â‚¹' symbol and commas
        cleaned_price = price.split('/')[0].replace('â‚¹', '').replace(',', '').strip()
        # Convert to float
        return float(cleaned_price)
    except (ValueError, AttributeError):
        # Return NaN if conversion fails
        return np.nan
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price)
```

```python
df['price'] = df['price'].astype(str).str.replace('â‚¹', '').str.replace(',', '').astype(float)
```

```python
df['area'] = round((df['price'] * 10000000) / df['price_per_sqft'])
```

```python
df.insert(loc=1,column='property_type',value='house')
```
